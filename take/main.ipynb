```python
import pandas as pd
import os
import openpyxl
```

```python
# Задаем путь к папке с Excel-файлами
directory = './'

# Указываем часть имени файлов, которые мы хотим обрабатывать
file_keyword = 'Справочник_'

# Создаем список для хранения данных
data = []

# Указываем имя листа, с которого будем считывать данные
sheet_name = 'Справочник проектов'  # Замените на название листа, если нужно

# Цикл по файлам в указанной папке
for filename in os.listdir(directory):
    if filename.endswith('.xlsx') and file_keyword in filename:
        file_path = os.path.join(directory, filename)

        # Чтение Excel-файла с указанием листа
        df_temp = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')

        # Извлечение значений из конкретных ячеек
        # D10 - это строка 9 и столбец 3 (0 индекс)
        # E10 - это строка 9 и столбец 4 (0 индекс)
        # G10 - это строка 9 и столбец 6 (0 индекс)
        # H10 - это строка 9 и столбец 7 (0 индекс)

        for i in range(0, 450):
            data.append({
                'objectname':df_temp.iloc[i, 0],
                'cod_class': df_temp.iloc[i, 1],  # G10
                'tip_object': df_temp.iloc[i, 2],  # H10
                'story': df_temp.iloc[i, 3],  # E10
                'napr': df_temp.iloc[i, 4],  # G10
                'div': df_temp.iloc[i, 5],  # H10
                'open': df_temp.iloc[i, 6],  # E10
                'priority': df_temp.iloc[i, 7],  # G10
                'opercom': df_temp.iloc[i, 8],  # H10
                'kur_da': df_temp.iloc[i, 9],  # E10
                'ocks_tz': df_temp.iloc[i, 10],  # G10
                'level': df_temp.iloc[i, 11],  # H10
                'fb': df_temp.iloc[i, 12],  # E10
                'gp': df_temp.iloc[i, 13],  # G10
                'np': df_temp.iloc[i, 14],  # H10
                'fp': df_temp.iloc[i, 15],  # E10
                'land': df_temp.iloc[i, 16],  # G10
                'region': df_temp.iloc[i, 17],  # H10
                'not_load': df_temp.iloc[i, 18],  # G10
                'kur': df_temp.iloc[i, 19],  # H10
                'zak': df_temp.iloc[i, 20],  # E10
                'jc': df_temp.iloc[i, 21],  # G10
            })

# Создание датафрейма из списка словарей
result_df = pd.DataFrame(data)

# Вывод полученного датафрейма
print(result_df)
```
