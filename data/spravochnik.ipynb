```python
file = '240710_1650_Справочник_живых_ОКС_по_уровням_ИТОГ_1.xlsx'
```

```python
list = [""] * 442
for i in range(442):
    list[i] = xl.iloc[i+2, 0]
    print(xl.iloc[i+2, 0])
```

```python
def remove_duplicates(array):
    """
    Удалить дубликаты из массива.

    Args:
        array (list): Исходный массив.

    Returns:
        list: Массив без дублей.
    """

    # Создать множество из исходного массива
    unique_set = set(array)

    return unique_set
```

```python
for i in range(len(list)):
    if list[i].find(" (") != -1:
        list[i] = list[i][0:list[i].find("(") - 1]
    if list[i].find(",") != -1:
        list[i] = list[i][0:list[i].find(",")]
    if list[i].find(" - ") != -1:
        list[i] = list[i][0:list[i].find(" - ")]
    if list[i].find("2025") != -1:
        list[i] = list[i][0:list[i].find(" 2025")]
    if list[i].find("эб") != -1:
        list[i] = list[i][0:list[i].find(" эб")]

projects = remove_duplicates(list)
pr = sorted(projects)
pr[0]
projects
```

```python
dict_projects = {}
dict_projects_full_name = {}

for i in range(len(projects)):
    dict_projects[pr[i]] = f'ИП №{i}'
    dict_projects_full_name[pr[i]] = f'Инвестиционный проект №{i}'
```

```python
dict_countries = {}
countries = [""] * 442

for i in range(442):
    countries[i] = xl.iloc[i+2, 17]
    t = xl.iloc[i+2, 0]
    for key in dict_projects.keys():
        if t.find(key) != -1:
            dict_countries[key] = countries[i]
dict_countries
```

```python
dict_ids = {}
ids = [""] * 442

for i in range(442):
    ids[i] = xl.iloc[i+2, 1]
    t = xl.iloc[i+2, 0]
    for key in dict_projects.keys():
        if t.find(key) != -1:
            temp0 = int(ids[i][0:1]) + random.randint(0, 5)
            temp = int(ids[i][2:3]) + random.randint(0, 4) 
            temp2 = int(ids[i][12:15]) + random.randint(100, 900)
            ids[i] = ids[i][0:2] + str(temp) + ids[i][3:10] + str(temp2)
            dict_ids[key] = ids[i]
dict_ids
```

```python
with open("output.txt", "w") as file:
    for key in dict_projects.keys():
        file.write(f"INSERT INTO investment_projects (id_project, full_name, short_name, country) values ('{dict_ids[key]}', '{dict_projects_full_name[key]}', '{dict_projects[key]}', '{dict_countries[key]}');\n")
print("Данные успешно записаны в файл!")
```

```python
# Открытие файла в режиме чтения ("r")
with open("input.txt", "r", encoding='utf-8') as file:
    # Создание пустого списка для хранения данных
    countries = []

    # Чтение файла построчно
    for line in file:
        # Удаление символа новой строки (\n) из каждой строки
        country = line.strip()
        # Добавление строки в массив
        countries.append(country)

# Вывод массива
print(countries)
```

```python
dict_id_countries = {}

for key in dict_countries.keys():
    try:
        dict_id_countries[key] = countries.index(dict_countries[key])
    except ValueError:
        dict_id_countries[key] = countries.index("Россия")
dict_id_countries
```
